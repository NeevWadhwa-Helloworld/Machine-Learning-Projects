## **Email spam detection with natural language model**

```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
import string
```

```python
df = pd.read_csv("/content/Requirement (2).zip")
df.head()
```

```python
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
nltk.download("stopwords")
```

```python
import nltk
from nltk.corpus import stopwords
import string

# Define stopwords set once for efficiency
STOPWORDS = set(stopwords.words('english'))

def process(text):
    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = ''.join(nopunc)

    clean = [word for word in nopunc.split() if word.lower() not in STOPWORDS]
    return clean
# to show the tokenization
df['text'].head().apply(process)
```

## **convert the text into a matrix of token counts**

```python
from sklearn.feature_extraction.text import CountVectorizer
message = CountVectorizer(analyzer=process).fit_transform(df['text'])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(message, df['spam'], test_size=0.20, random_state=0)
print(message.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(xtrain, ytrain)
```

## **see the clssifiers prediction and actual values on the data set**

```python
print(classifier.predict(xtrain))
print(ytrain.values)
```

```python
# Evaluating the model on the training data set
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(xtrain)
print(classification_report(ytrain, pred))
print()
print("Confusion Matrix: \n", confusion_matrix(ytrain, pred))
print("Accuracy: \n", accuracy_score(ytrain, pred))
```

```python
print(classifier.predict(xtest))
print(ytest.values)
```

## **evaluate**

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(xtest)
print(classification_report(ytest, pred))
print()
print("Confusion Matrix: \n", confusion_matrix(ytest, pred))
print("Accuracy: \n", accuracy_score(ytest, pred))
```
